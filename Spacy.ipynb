```python
import spacy
```

```python
nlp = spacy.blank("en")

doc = nlp('''Dr. Strange loves pav bhaji of mumbai as it costs only 2$ per plate.
            tasi@gmail.com and ram_@outlook.com gimme more emails ''')

for tokens in doc:
    print(tokens)
```

```python
emails = []
for token in doc:
    if token.like_email:
        emails.append(token.text)
        
        
```

```python
print(emails)
```

```python
from spacy.symbols import ORTH

nlp.tokenizer.add_special_case("gimme",[
    {ORTH:"gim"},
    {ORTH:"me"}
])


tokens = [token.text for token in doc]
tokens
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
words = ["eating", "jugging", "ate", "ability", "meeting"]

for word in words:
    print(word," | " ,stemmer.stem(word))
```

```python
nlp = spacy.load("en_core_web_sm")

ar = nlp.get_pipe('attribute_ruler')

ar.add([[{"TEXT":"Bro"}],[{"TEXT":"Brah"}],[{"TEXT":"bruh"}]],{"LEMMA":"Brother"})

doc = nlp("eating eats etc eat ate adjustable rafting ability meeting better Bro bruh Brah")
doc_fil = []

for token in doc:
    if token.pos_ not in ["SPACE","X","PUNCT"]:
        doc_fil.append(token)
        
for token in doc_fil:
    print(token.text," | ",token.lemma_, " | ",token.pos_)
```

```python
doc = nlp("Tesla is going to acquire Twitter for $45 billion")

for ent in doc.ents:
    print(ent.text," | ", ent.label_," | ",spacy.explain(ent.label_))
    
```

```python
from spacy.tokens import Span

s1 = Span(doc,0,1,label="ORG")
s2 = Span(doc,5,6,label="ORG")

doc.set_ents([s1, s2],default="unmodified")

for ent in doc.ents:
    print(ent.text," | ", ent.label_," | ",spacy.explain(ent.label_))
```
